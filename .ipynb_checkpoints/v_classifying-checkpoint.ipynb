## MARTIN's Shape index
This seems to indentify pretty well the first split the geometries between the odd and the 'normal'/ expected

```python
import pygeos
```

```python
# Polygonize street network
polygons = pygeos.polygonize(mad.geometry.array.data)

# Store geometries as a GeoDataFrame
polygons = gpd.GeoDataFrame(
    geometry=gpd.GeoSeries([polygons], crs=mad.crs).explode(ignore_index=True)
    )
polygons.tail(2)
```

```python
# Extract underlying PyGEOS geometries pygeos understands (minimum_bounding_circle is not yet exposed in geopandas)
ga = polygons.geometry.array.data
ga
```

```python
# measure area
polygons["area"] = polygons.area
# generate minimum bounding circle
circles = gpd.GeoDataFrame(
    geometry=gpd.GeoSeries(pygeos.minimum_bounding_circle(ga), crs=polygons.crs))
# measure MBC's area
circles["circle_area"] = circles.area
# measure Reock (circular) compactness
polygons["reock"] = polygons["area"] / circles["circle_area"]
# measure direct shape index that captures banana-like relationship between area and Reock compactness in a one dimension
polygons["shape_index"] = polygons["area"] / np.sqrt(circles["circle_area"])
```

```python
polygons.tail()
```

```python
len(polygons) == len(mad_polys)
```

```python
polygons.shape_index.plot.hist(figsize = (10,5),bins = 50);
```

```python
polygons[polygons.shape_index == polygons.shape_index.max() ].plot();
```

```python
s = (polygons.area <= 50000) & (polygons.reock < 0.7)
polygons[s].plot.scatter(x= 'area', y = 'reock',
                      alpha = 0.3, marker = '.', c = 'shape_index',
                      colormap='viridis',
                      figsize = (12,6)) ;
```

```python
polygons[s].shape_index.plot.hist(bins = 80, edgecolor ='white', figsize = (12,6))
plt.axvline(25, color='red', linewidth=2) ;
```

```python
polygons['valid_poly'] = (polygons.shape_index > 25) #| (polygons.reock > 0.7) #establishing a first break

s = (polygons.area <= 50000)
polygons[s].plot.scatter(x= 'area', y = 'reock',
                      alpha = 0.25 , marker = '.', c = [ 'green' if i else 'red' for i in polygons[s].valid_poly]    
                     ,figsize = (12,6)) ;
```
